```python
import tensorflow as tf 
import efficientnet.keras as efn
import cv2
import numpy as np 
```

```python
model = tf.keras.models.load_model('model.h5')
```

```python
model.summary()
```

```python
# Read Image
def read_image(path, im_size, normalize_image = False):
    img = cv2.imread(path, cv2.IMREAD_COLOR)  
    img = cv2.resize(img, (im_size, im_size))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB).astype(np.float32)
    if normalize_image:
        img /= 255.0  
    return img
```

```python
def create_image_tta(im_res):
    # Flip Image Left-Right
    im_res_lr = np.fliplr(im_res)

    return np.stack((im_res, im_res_lr))
```

```python
file_name='test.jpg'
SIZE = 256
image = read_image(file_name, SIZE, normalize_image = True)
```

```python
image.shape
```

```python
result = model.predict(create_image_tta(image))
```

```python
max_value = np.max(np.mean(result, axis = 0))
max_index = np.argmax(np.mean(result, axis = 0))
```
